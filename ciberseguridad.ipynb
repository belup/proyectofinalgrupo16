TEMA: Ciberseguridad. Ataques cibernéticos.

Los ataques cibernéticos son acciones maliciosas dirigidas a sistemas informáticos, redes o dispositivos con el objetivo de robar, alterar o destruir información. Entre los más comunes se encuentran el phishing (engaño para obtener datos personales), malware (software malicioso como virus o ransomware), ataques DDoS (saturación de servidores para interrumpir servicios) y exploits (aprovechamiento de vulnerabilidades del sistema). Estos ataques pueden afectar tanto a individuos como a empresas y gobiernos, causando pérdidas económicas, robo de identidad y daño a la reputación.


DATASET: Descargado de Kaggel, https://www.kaggle.com/datasets/teamincribo/cyber-security-attacks/data

El objetivo es ver que tipo de ataques ciberneticos, el nivel que tuvo, en que zonas de la region y las acciones tomadas al respecto.

```python
#renombramos las columnas
ciberseguridad_seleccionadas.rename(columns={'Timestamp': 'Fecha y hora', 'Source IP Address': 'Dirección IP origen', 'Destination IP Address': 'Dirección IP destino', 'Attack Type': 'Tipo de ataque', 'Action Taken': 'Acción tomada', 'Severity Level': 'Nivel de gravedad', 'Geo-location Data': 'Datos de geolocalización'}, inplace=True)
```

```python
#llamamos al dataset para verlo
ciberseguridad_seleccionadas
```

```python
ciberseguridad_seleccionadas["Nivel de gravedad"].value_counts()
#vemos que nivel de gravedad tienen los ataques
```

```python
#Vemos que atributos únicos hay en la columna Nivel de gravedad
ciberseguridad_seleccionadas["Nivel de gravedad"].unique()
```

```python
#reemplazamos los nombres de los valores de la columna Nivel de gravedad
ciberseguridad_seleccionadas['Nivel de gravedad'] = ciberseguridad_seleccionadas['Nivel de gravedad'].str.replace('Low', 'c').str.replace('Medium','B').str.replace('High','A', regex=False).astype(str)
```

```python
#contamos cuantos valores hay en cada atributo unico
ciberseguridad_seleccionadas["Nivel de gravedad"].value_counts()
```

```python
#Visualizamos los primeros diez valores
ciberseguridad_seleccionadas.head(10)
```

```python
# Graficamos la columna Tipo de ataque para visualizar la cantidad de valores unicos

from matplotlib import pyplot as plt
import seaborn as sns
ciberseguridad_seleccionadas.groupby('Tipo de ataque').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#Graficamos el tipo de ataque
plt.figure()
plt.hist(ciberseguridad_seleccionadas['Tipo de ataque'].dropna(), bins=10, color="blue", alpha=0.3)
plt.ylabel("Nivel de gravedad", size=14)
plt.xlabel("Tipo de ataque", size=14)
plt.title("Nivel de gravedad de los ataques ciberneticos", size=16, pad=25)
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)
plt.grid(True)
plt.show()
```

```python
#Graficamos el nivel de gravedad segun el tipo de ataque
plt.figure()
sns.displot(ciberseguridad_seleccionadas["Tipo de ataque"], color="#5ea88e", kde=True)
plt.xlabel('Tipo de ataque')
plt.ylabel('Nivel de gravedad')
plt.title('Nivel de gravedad de los ataques ciberneticos  por tipo de ataque')
sns.despine()
plt.show()
```
